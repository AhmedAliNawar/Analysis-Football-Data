```python
player = data["Player Names"].value_counts()== 5
TopPlayer = player[player].index
TopPlayer
```

```python
DataTopPlayer = data[data["Player Names"].isin(TopPlayer)]
DataTopPlayer.head()
```

```python
df_top = DataTopPlayer.groupby(by='Player Names')["Matches_Played", "Goals", "Mins", "OnTarget"].sum().reset_index()
df_top.head()
```

```python
plt.figure(figsize=(10, 6))
TopByGoals = df_top.sort_values(by=['Goals'], ascending=False)
sns.barplot(x='Goals', y='Player Names', data=TopByGoals)
plt.show()
```
